```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

This command is use to read the file.


```python
df=pd.read_csv('/jobs.csv')
df.head(20)
```

The command in pandas provides a summary of the Dataset.


```python
df.info()
```

The command in pandas provides summary statistics for numerical columns in a Dataset.


```python
df.describe()
```

This command is used  to display the duplicates in the dataset.

```python
display("Duplicates:", df.duplicated())
```

This command is used to count the number of missing values.

```python
display(df.isnull().sum())
```

This command is used to remove rows with the missing values.

```python
df_dropna = df.dropna()
display(df_dropna)
```

This command is used to remove column with the missing values.

```python
df_dropna_cols = df.dropna(axis=1)
display(df_dropna_cols.head())
```

This command is used to select the  employee having months_experience more than 50.

```python
df_loc_1 = df.loc[df['months_experience'] > 50]
display(df_loc_1.head())
```

```python
df_loc_2 = df.loc[(df['company'] == 'Apple') ]
display(df_loc_2.head(20))
```

```python
df_iloc_1 = df.iloc[:5]
display(df_iloc_1)
```

 Select rows 10 to 15 and columns 2 to 5

```python
df_iloc_2 = df.iloc[10:15, 2:6]
display(df_iloc_2)
```

```python
df_query_1 = df.query("post_id > 26000000")
display(df_query_1.head())
```

```python

df_query_1 = df.query("months_experience > 60")
display(df_query_1.head())
```

```python
df_query_2 = df.query('title == "Senior Data Engineer"')
display(df_query_2.head())
```

```python
df_between_1 = df[df['months_experience'].between(60, 80)]
display(df_between_1)
```

```python
df_mask_1 = df.copy()
df_mask_1['months_experienc'] = df_mask_1['months_experience'].mask(df_mask_1['months_experience'] > 50, 'NOT ELIGIBLE')
display(df_mask_1.head(30))
```

```python
df_groupby = df.groupby('Seniority level')['months_experience'].mean()
display(df_groupby)
```

```python

df_groupby_multi = df.groupby(['Seniority level', 'company'])[['months_experience', 'sal_low']].mean()
display(df_groupby_multi)
```

```python

df_agg = df.groupby('Job function').agg({'sal_high': ['mean', 'max', 'min'], 'sal_low': ['mean', 'max', 'min']})
display(df_agg)
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df['months_experience'])
plt.show()
```

```python
sns.countplot(x='months_experience', data=df)
plt.title("experienced")
plt.show()
```
